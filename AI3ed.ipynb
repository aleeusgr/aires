```python
print('semantic net representation')
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.name = 'recursion'
G.add_nodes_from([1,2])
G.add_edge(1,2)
nx.set_edge_attributes(G,{(1,2):{'type':'initiate'}})
nx.set_node_attributes(G,{1: {"type": "MS"}, 
                          2: {"type": "MS"}})

plt.figure(figsize=(8,8))
plt.axis('off')

# generate the layout and place nodes and edges
layout = nx.circular_layout(G)

# plot nodes, labels, and edges with options
nx.draw_networkx_nodes(G, pos=layout, node_size=500, alpha=0.8)
nx.draw_networkx_edges(G, pos=layout, width=3, style='dotted',
                       edge_color='orange')
nx.draw_networkx_labels(G,labels=nx.get_node_attributes(G,'type'), pos=layout, font_size=15)
nx.draw_networkx_edge_labels(G,labels=nx.get_edge_attributes(G,'type'), pos=layout, font_size=15)

plt.show()
```
